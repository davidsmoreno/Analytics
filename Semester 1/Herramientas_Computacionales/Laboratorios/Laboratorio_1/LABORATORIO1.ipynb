## Fase 1: obtener e inspeccionar archivos

Los datos de encuestas fueron obtenidos como parte de un trabajo de campo y, por limitaciones tecnológicas, solo se cuenta con la información en archivos de texto plano (`.txt`). Los archivos con los que cuentas son:

* `"genero.txt"`: un archivo con el género de cada persona encuestada;

* `"region.txt"`: un archivo con la región de proveniencia de cada persona encuestada.

### Misión 1

**Versión base**: utiliza métodos de lectura de archivos para guardar la información de cada archivo en una estructura de datos de Python, de manera que se respete el orden y la categoría de los datos.

**Versión completa**: escribe una función que sea capaz de ejecutar la **versión base** y llámala dentro de un ciclo para automatizar la lectura de los archivos.

```python

directorio = "./Archivos/"
Archivos=["edad.txt","escolaridad.txt","estado_civil.txt","estrato.txt","genero.txt","promedio.txt","region.txt"]


def read_txt(directorio):
    with open(directorio, "r") as f:
        lines = f.readlines()

    cleaned_lines = [line.strip() for line in lines]

    return cleaned_lines

def version_completa():
    matriz={}
    for i in Archivos:
        file_path=directorio+str(i)
        content=read_txt(file_path)
        matriz[i]=content
    matriz["edad.txt"]=[int(i) for i in matriz["edad.txt"]]
    matriz["promedio.txt"]=[float(i) for i in matriz["promedio.txt"]] #Parse the data to int and float for the numeric values
    return matriz

#Creamos un diccionario con los datos de los archivos donde la llame es el nombre de la columna
datos=version_completa()
datos
```

### Misión 2

**Versión base**: calcula estadísticas descriptivas básicas (por lo menos, media/mediana y varianza/desviación) para cada variable relevante en el contexto del problema y repórtalas concluyendo acerca de las principales características de los datos.

**Versión completa**: complementa la **versión base** con visualizaciones informativas sobre los datos (ej., histogramas, gráficos de dispersión, boxplots, etc.), relevantes para el problema.

```python
# Implementa tu respuesta en esta celda
## Estadisticas descripitivas
def media_valor(x):
    return sum(x)/len(x)

def varianza(x):
    mu=media_valor(x)
    return sum([(xi-mu)**2 for xi in x])/len(x)

def mediana(x):
    n=len(x)
    x_ordenado=sorted(x)
    if n%2==0:
        return (x_ordenado[n//2-1]+x_ordenado[n//2])/2
    else:
        return x_ordenado[n//2]
    
def desviacion_estandar(x):
    return varianza(x)**0.5

def estadisticas_descriptivas(x):
    return media_valor(x),varianza(x),mediana(x),desviacion_estandar(x)

#En el contexto del problema solo tenemos 2 variables numericas, las cuales son edad y promedio


edad_estadisticas=estadisticas_descriptivas(datos["edad.txt"])
promedio_estadisticas=estadisticas_descriptivas(datos["promedio.txt"])

print(edad_estadisticas)
print(promedio_estadisticas)





```

### Misión 4

**Versión base**: desarrolla un algoritmo que seleccione a los potenciales beneficiarios de las becas, según la primera propuesta.

**Versión completa**: escribe una función que sea capaz de ejecutar la **versión base** para distintos números $N$ de becas disponibles.

```python
# Implementa tu respuesta en esta celda

def asignar_becas_p1(datos,N_becas):
    estudiantes = [{column: datos[column][i] for column in Archivos} for i in range(len(datos["edad.txt"]))]
    estudiantes_ordenados=sorted(estudiantes,key=lambda x: x['promedio.txt'],reverse=True) #Mayor a menor
    becas_asignadas={}
    beneficiarios=[]
    for estudiante in estudiantes_ordenados:
        genero=estudiante['genero.txt']
        region=estudiante['region.txt']

        if (genero,region) not in becas_asignadas:
            becas_asignadas[(genero,region)]=0
        if becas_asignadas[(genero,region)]<N:
            beneficiarios.append(estudiante)
            becas_asignadas[(genero,region)]+=1
        if len(beneficiarios)==N_becas:
            break
    return beneficiarios



N = 10  # Cambia este valor según la cantidad de becas disponibles
beneficiarios_propuesta_1 = asignar_becas_p1(datos, N)
beneficiarios_propuesta_1


```

### Misión 5

**Versión base**: desarrolla un algoritmo que seleccione a los potenciales beneficiarios de las becas, según la segunda propuesta.

**Versión completa**: escribe una función que sea capaz de ejecutar la **versión base** para distintos números $N$ de becas disponibles.

```python
# Implementa tu respuesta en esta celda
def asignar_becas_p2(datos, N_becas):
    estudiantes = [{column: datos[column][i] for column in Archivos} for i in range(len(datos["edad.txt"]))]    
    # Crear una lista de estratos únicos y ordenarlos en orden ascendente
    estratos_unicos = sorted(list(set(datos['estrato.txt'])))
    
    becas_asignadas = set()
    beneficiarios = []
    
    for estrato in estratos_unicos:
        # Filtrar estudiantes solo para el estrato actual
        estudiantes_estrato = [estudiante for estudiante in estudiantes if estudiante['estrato.txt'] == estrato]
        
        # Ordenar los estudiantes del estrato actual en orden descendente de promedio
        estudiantes_estrato = sorted(estudiantes_estrato, key=lambda x: x['promedio.txt'], reverse=True)
        
        # Calcular el número de becas disponibles para este estrato (2% del total)
        becas_disponibles_estrato = int(len(estudiantes_estrato) * 0.02)
        
        if becas_disponibles_estrato > 0:
            for estudiante in estudiantes_estrato[:becas_disponibles_estrato]:
                beneficiarios.append(estudiante)
                becas_asignadas.add(tuple(estudiante.items()))
        
        if len(becas_asignadas) >= N_becas:
            break
    
    return beneficiarios

N = 10  # Cambia este valor según la cantidad de becas disponibles
beneficiarios_propuesta_2 = asignar_becas_p2(datos, N)
beneficiarios_propuesta_2

```

### Misión 7

Desarrolla un algoritmo que reciba los parámetros para configurar la nueva propuesta y seleccione a los potenciales beneficiarios de las becas.

```python
# Implementa tu respuesta en esta celda
def asignar_becas_propuesta(datos, N, min_edad, max_edad, caracteristicas, porcentajes):
    # Crear una lista de diccionarios para representar a los estudiantes
    estudiantes = [dict(zip(datos.keys(), values)) for values in zip(*datos.values())]

    # Filtrar por edad
    datos_filtrados = [estudiante for estudiante in estudiantes if min_edad <= estudiante['edad.txt'] <= max_edad]

    # Crear un diccionario para almacenar los potenciales beneficiarios por combinaciones de características
    beneficiarios_por_combinacion = {}

    # Agrupar potenciales beneficiarios por combinaciones de características
    for estudiante in datos_filtrados:
        clave = tuple(estudiante[caracteristica] for caracteristica in caracteristicas)
        if clave not in beneficiarios_por_combinacion:
            beneficiarios_por_combinacion[clave] = []
        beneficiarios_por_combinacion[clave].append(estudiante)

    # Asignar becas basadas en el mejor promedio académico en cada combinación de características
    beneficiarios_seleccionados = []

    for combinacion in porcentajes.keys():
        porcentaje_total = porcentajes[combinacion]
        grupos = beneficiarios_por_combinacion.keys()
        
        for grupo in grupos:
            estudiantes_en_grupo = beneficiarios_por_combinacion[grupo]
            estudiantes_en_grupo = sorted(estudiantes_en_grupo, key=lambda x: x['promedio.txt'], reverse=True)
            porcentaje = porcentajes[combinacion] / len(grupos)
            becas_asignar = int(N * porcentaje_total * porcentaje / 100)
            becas_asignar = min(becas_asignar, len(estudiantes_en_grupo))
            beneficiarios_seleccionados.extend(estudiantes_en_grupo[:becas_asignar])

    return beneficiarios_seleccionados




# Ejecuta la función con la configuración y otros parámetros
```

## Ventajas de la Tercera Propuesta:

1. **Mayor Granularidad:** Esta propuesta permite una mayor granularidad en la asignación de becas al considerar tres características: género, estrato y escolaridad. Esto significa que se pueden considerar de manera más específica las necesidades y condiciones de diferentes grupos dentro de la población.

2. **Equidad en la Asignación:** Los porcentajes definidos aseguran que las becas se asignen de manera equitativa entre diferentes grupos, lo que puede promover la igualdad de oportunidades. Esto es especialmente importante en contextos donde existen disparidades significativas entre grupos.

3. **Personalización:** Al tener en cuenta múltiples características, se puede personalizar la asignación de becas de acuerdo con las necesidades y características de los solicitantes. Por ejemplo, se pueden asignar más becas a mujeres con educación secundaria si se identifica que este grupo enfrenta desafíos específicos.

4. **Transparencia:** La asignación de becas se basa en porcentajes definidos claramente, lo que hace que el proceso sea transparente y comprensible para todos los involucrados, incluidos los solicitantes y el comité de selección.

## Desventajas de la Tercera Propuesta:

1. **Complejidad:** La inclusión de múltiples características y combinaciones puede hacer que el proceso de asignación de becas sea más complejo y requerir un cálculo más detallado. Esto puede aumentar la carga de trabajo administrativa.

2. **Posibilidad de Desigualdades Ocultas:** Aunque se busca la equidad, existe la posibilidad de que algunas combinaciones de características reciban más atención o recursos que otras, lo que podría generar desigualdades ocultas.



4. **Posible Subjetividad:** La asignación basada en múltiples características puede ser subjetiva si no se definen criterios claros para determinar cuál es la combinación más adecuada para recibir una beca.

En comparación con las propuestas anteriores:

- La primera propuesta se basa únicamente en género y región, lo que puede simplificar el proceso pero no garantiza la equidad total.
- La segunda propuesta se centra en género y estrato, lo que proporciona una mayor equidad que la primera propuesta pero es menos granular que la tercera.

En última instancia, la elección de la propuesta dependerá de los objetivos específicos de asignación de becas, la disponibilidad de recursos y la importancia de considerar múltiples características en la toma de decisiones.
