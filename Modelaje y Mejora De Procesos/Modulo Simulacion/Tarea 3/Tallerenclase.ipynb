## Genera en python una secuencia de 10000 númros aleatorios a través del método de congruencia lineal

Queremos crear la siguiente secuencia de números en python. Para esto vamos a escribir la formula $X_{i+1}=(aX_{i}+c) mod m, i=0,1,2,3,...$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def secuancia_pseudoaleatoria(x_0,a,c,m,n):
    
    Secuencia=[]
    for i in range(n):
        x_i=((a*x_0+c)%m)/float(m)
        Secuencia.append(x_i)
        x_0=x_i
    return Secuencia

##
x_0 = 2020
m = 2**31+3
a = 2**16+1
c = 0
n = 10000

sim_1=secuancia_pseudoaleatoria(x_0,a,c,m,n)


# Create a histogram with more bins for better visualization
plt.hist(sim_1, bins=50, edgecolor='black', color='blue',range=(0,1))
plt.xlabel('Generated Values')
plt.ylabel('Frequency')
plt.title('Histogram of Pseudo-Random Numbers')
plt.grid(True)
plt.show()

```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf
autocorrelation = acf(sim_1)

plt.figure(figsize=(10, 5))
sm.graphics.tsa.plot_acf(autocorrelation, lags=20, title='Autocorrelation Plot')
plt.show()
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución exponencial
lambda_param = 0.5  # Puedes ajustar este valor según tus necesidades

# Valores de x para los que deseas calcular la función acumulada
x_values = np.linspace(0, 5, 100)  # Valores de 0 a 5 en 100 pasos

# Calcula la función acumulada para cada valor de x
cumulative_probs = scipy.stats.expon.cdf(x_values, scale=1/lambda_param)

# Grafica la función acumulada
plt.plot(x_values, cumulative_probs)
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.title('Función Acumulada de Distribución Exponencial')
plt.grid(True)
plt.show()
cumulative_probs

```
