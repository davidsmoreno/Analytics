## Clase 2 - Análisis de entrada

Generación de muestras de una variable aleatoria normal usando numpy.

```python
import pandas as pd

df = pd.read_csv('mmps_clase2_datos.csv')
df
```

```python
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=df['TiempoEntreArribos'], bins=30)
plt.title('Histograma de tiempos de arribo')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

# Bondad de ajuste

Densidad normal estándar.

```python
from distfit import distfit
```

```python
dfit = distfit(todf=True)
dfit.fit_transform(df['TiempoEntreArribos'])
dfit.plot()
```

```python
dfit.plot_summary()
plt.axhline(y = 0.05, color = 'r', linestyle = '-')
plt.show()
```
