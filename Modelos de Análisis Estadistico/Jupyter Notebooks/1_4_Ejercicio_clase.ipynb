```python
import seaborn as sns

sns.set(style="ticks")
sns.pairplot(df, hue="Species")

```

Observando estas distribuciones, ¬øcu√°n dif√≠cil crees que ser√° clasificar a los ping√ºinos utilizando s√≥lo la "profundidad del culmen" y la "longitud del culmen"?

```python
# Create a FacetGrid to visualize distribution of each feature

features_df = df.drop(columns=['Species'])
g = sns.FacetGrid(features_df, col="island", hue="island", col_wrap=3)
g.map_dataframe(sns.histplot)
g.add_legend()
plt.show()
```
