```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("adult-census.csv")

# Plot the distribution of numerical variables
numerical_columns = df.select_dtypes(include=['int64'])
plt.figure(figsize=(12, 6))
for col in numerical_columns.columns:
    sns.histplot(data=numerical_columns, x=col, bins=20, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Plot the correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object'])

# Explore distribution of each categorical variable
for col in categorical_columns.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Explore relationships between categorical and numerical variables
for col in categorical_columns.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=col, y='age')
    plt.title(f'Relationship between {col} and Age')
    plt.xlabel(col)
    plt.ylabel('Age')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Select a categorical variable and a numerical variable
categorical_variable = "education"
numerical_variable = "age"

# Create a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x=categorical_variable, y=numerical_variable)
plt.title(f'Boxplot of {numerical_variable} by {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel(numerical_variable)
plt.xticks(rotation=45)
plt.show()
```

```python
categorical_variable = "education"
numerical_variable = "capital-gain"

# Calculate the mean of the numerical variable for each category
mean_per_category = df.groupby(categorical_variable)[numerical_variable].mean()

# Calculate the total mean of the numerical variable
total_mean = df[numerical_variable].mean()

# Calculate the mean as a percentage of the total mean
mean_per_category_percent = (mean_per_category / total_mean) * 100

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_per_category_percent.index, y=mean_per_category_percent.values)
plt.title(f'Mean {numerical_variable} as Percentage by {categorical_variable}')
plt.xlabel(categorical_variable)
plt.ylabel(f'Mean {numerical_variable} (%)')
plt.xticks(rotation=45)
plt.show()
```
